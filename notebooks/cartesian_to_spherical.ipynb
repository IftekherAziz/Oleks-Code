## Transformation Functions

To obtain the spherical coordinates of each star from its Cartesian coordinates, I used following formulae: 
$$ r = \sqrt{x^2 + y^2 + z^2} $$
$$ \theta = \arccos{\frac{z}{r}} $$
$$ \phi = \operatorname{atan2}(y, x) $$
where $r$ is the radial distance, $ \theta$ is the polar angle and $\phi$ is the azimuthal angle.

Declination $\delta$ is obtained by
$$ \delta = 90^{\circ} - \theta$$
Right ascension $\alpha$ is simply the azimuthal angle
$$ \alpha = \phi $$

To transform Cartesian velocity vectors into spherical vectors I used following matrix:
$$ T=
\begin{pmatrix}
    \sin(\theta)\cos(\phi) & \sin(\theta)\sin(\phi) & \cos(\theta) \\
    \cos(\theta)\cos(phi) & \cos(\theta)\sin(\phi) & -\sin(\theta) \\
    -\sin(\phi) & \cos(\phi) & 0 \\
\end{pmatrix}$$


Obtained vectors $\hat{\theta}$, $\hat{\phi}$ give us the tangential velocities of the given points in spherical coordinate system. Now, to obtain the angular velocities, also defined as $\mu_ {\alpha} $ (right ascension) and $\mu_\delta$ (declination), I divided them by distance $r$. In the case of $\mu_\alpha$ there is also a correction factor $\cos(\delta) = \sin(\theta)$ that accounts for changes of the radius of the sphere on different latitudes.
